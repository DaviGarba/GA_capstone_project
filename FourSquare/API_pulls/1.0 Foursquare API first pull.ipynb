### Variables

```python
def extracting_variables(results):
    venue_dict = {}
    
    venue_name = []
    venue_id = []
    rating = []
    cat_id = []
    cat_name = []
    ll = []
    lng = []
    lat = []

    for item in results:

        try: 
            venue_name.append(item['venue']['name'])
            venue_id.append(item['venue']['id'])
#             rating.append(item['venue']['rating'])
            cat_name.append(item['venue']['categories'][0]['name'])
            cat_id.append(item['venue']['categories'][0]['id'])
            ll.append(str(item['venue']['location']['lat']) + ',' + str(item['venue']['location']['lng']))
            lat.append(item['venue']['location']['lat'])
            lng.append(item['venue']['location']['lng'])
            
        except:
            venue_name.append(np.nan)
            venue_id.append(np.nan)
#             rating.append(np.nan)
            cat_name.append(np.nan)
            cat_id.append(np.nan)
            ll.append(np.nan)
            lat.append(np.nan)
            lng.append(np.nan)


   
    venue_dict = {'venue_name': venue_name, 'venue_id':venue_id,'rating':rating, 'cat_name':cat_name,'cat_id':cat_id,'ll':ll, 'lng':lng, 'lat':lat}
   
#     return len(venue_name), len(venue_id), len(rating), len(cat_name), len(cat_id), len(ll)

    return venue_dict



# print len(venue_name)
# print len(venue_id)  
# print len(rating)
# print len(cat_name)
# print len(cat_id)
# print len(ll)


# tips = []
# for d in data:
#     for item in d['response']['groups'][0]['items']:
#         try:
#             tips.append(item['tips'][0]['text'])
#         except:
#             tips.append(np.nan)
# print len(tips)
```
