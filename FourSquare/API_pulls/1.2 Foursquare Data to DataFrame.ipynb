## Data to Results
 - data from each param converted into list of all venues


```python
def json_to_venue_list(list_):
    '''raw json data to list of entries (data for each venue) '''
    results = []
    for data in list_:
        for item in data:
            results.append(item['response'])
            
    return results


test_result = json_to_venue_list(test_open)
print len(test_result)
test_result[:10]
```
